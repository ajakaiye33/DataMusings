### DataFrames and BeautifulSoup Ex 5.11

```python
with open('./The-Data-Wrangling-Workshop/Chapter05/datasets/test.html','r') as fd:
    soup = BeautifulSoup(fd)
    data = soup.find_all('tr')
    print(f'Data is a {type(data)} and {len(data)} items long')
    data_without_header = data[1:]
    headers = data[0]
    print(headers)
    col_header = [th.get_text() for th in headers.find_all('th')]
    df_data = [[td.get_text() for td in tr.find_all('td')] for tr in data_without_header]
    print(df_data)
df = pd.DataFrame(df_data,columns=col_header)
df.head()
```

### Stacking URLs from a Documenting Using bs4 Ex 5.13

```python
with open('./The-Data-Wrangling-Workshop/Chapter05/datasets/test.html','r') as fd:
    soup = BeautifulSoup(fd)
    lis = soup.find('ul').find_all('li')
    stack = []
    for li in lis:
        a = li.find('a',href=True)
    stack.append(a['href'])
print(stack)
```

### Reading Tabular Data from a web page and Creating DataFrames Activity 5.01

```python
from urllib.request import urlopen
from urllib.error import HTTPError
from urllib.error import URLError
from bs4 import BeautifulSoup
import requests
```

```python
def get_site_html(url):
    try:
        html = requests.get(url)
    except HTTPError as e:
        return None
    try:
        bs = BeautifulSoup(html.content, 'html.parser')
        title = bs
    except AttributeError as e:
        return None
    return title
```

```python
result_html = get_site_html('https://en.wikipedia.org/wiki/List_of_countries_by_GDP_(nominal)')
grab_table = result_html.find_all('table',{'class':'wikitable sortable'})
```

```python
imf_table = grab_table[0]
w_Bnk_table = grab_table[1]
un_table = grab_table[2]
```

```python
# workign on the imf table
the_total = imf_table.find_all('tr')
the_headers = the_total[0]
other_parts = the_total[1:]
#get a clean columns
clean_head = [i.get_text().replace('\n','') for i in the_headers.find_all('th')]
print(clean_head)
```

```python
clean_body = [[td.get_text().replace('\xa0','').replace('\n','').replace('[22][n 1]','').replace('[n 2]','').replace('[19]','') for td in tr.find_all('td')] for tr in other_parts]
print(clean_body)
    
```

```python
# read with pandas dataframe
imf_df = pd.DataFrame(data=clean_body, columns=clean_head)
```

```python
imf_df
```

```python
def clean_country_col(entry):
    return entry.replace(',', '')
```

```python
imf_df['gdp_us_million'] = imf_df['GDP(US$million)'].apply(clean_country_col)
```

```python
imf_df.drop('GDP(US$million)', axis=1, inplace=True)
```
